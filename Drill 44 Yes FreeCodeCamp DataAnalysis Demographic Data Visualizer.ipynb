```python
paid_group = paid.groupby(["native-country"]).count()
paid_group
```

```python
topearner = df[["salary","native-country"]][df["salary"]==">50K"]
top_earner_group = topearner.groupby(["native-country"]).count()
top_earner_group
```

```python
the_top_percent = (top_earner_group/paid_group)*100
the_top_percent
```

```python
the_top_percent.sort_values(by="salary", ascending = False, inplace=True)

the_top_percent.index[0]
```
